```python
GOAL = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 0)
)

def possible_move(x, y):
    moves = []
    if x > 0: moves.append((-1, 0)) # Lên
    if x < 2: moves.append((1, 0))  # Xuống
    if y > 0: moves.append((0, -1)) # Trái
    if y < 2: moves.append((0, 1))  # Phải
    return moves

def find_zero(mat):
    for i in range(3):
        for j in range(3):
            if mat[i][j] == 0:
                return i, j

def simulate_move(mat, x, y, dx, dy):
    new_mat = [list(row) for row in mat]
    new_x, new_y = x + dx, y + dy
    new_mat[x][y], new_mat[new_x][new_y] = new_mat[new_x][new_y], new_mat[x][y]
    return tuple(tuple(row) for row in new_mat)

def print_mat(mat):
    for row in mat:
        print(list(row))

def bfs_8_puzzle(initial_state):
    frontier = [(initial_state, [])]

    reached = set()
    reached.add(initial_state)

    if initial_state == GOAL:
        return []

    while frontier:
        current_state, path = frontier.pop(0)

        x, y = find_zero(current_state)

        for dx, dy in possible_move(x, y):
            child_state = simulate_move(current_state, x, y, dx, dy)

            if child_state not in reached:
                new_path = path + [child_state]

                if child_state == GOAL:
                    return new_path

                reached.add(child_state)
                frontier.append((child_state, new_path))

    return None

def main():
    mat = (
        (1, 6, 0),
        (5, 3, 2),
        (8, 4, 7)
    )

    print("Bảng ban đầu:")
    print_mat(mat)

    solution = bfs_8_puzzle(mat)

    if solution is not None:
        print(f"\nThành công sau {len(solution)} bước.")
        for i, step_mat in enumerate(solution):
            print(f"\nB{i + 1}:")
            print_mat(step_mat)
    else:
        print("\nKhông tìm thấy lời giải!")

if __name__ == "__main__":
    main()
```
