```python
import random

row, col = random.randint(3, 6), random.randint(3, 6)

mat = []
for i in range(row):
    mat.append([random.randint(0, 2) for j in range(col)])

def possible_move(x, y, mat):
    move = []
    if x > 0 and mat[x-1][y] != 2:
        move.append((-1, 0)) # Lên
    if x < row-1 and mat[x+1][y] != 2:
        move.append((1, 0))  # Xuống
    if y > 0 and mat[x][y-1] != 2:
        move.append((0, -1)) # Trái
    if y < col-1 and mat[x][y+1] != 2:
        move.append((0, 1))  # Phải
    return move

def check_floor(mat):
    for i in range(row):
        for j in range(col):
            if mat[i][j] == 1:
                return False
    return True

def main():
    count = 0
    visited = set()

    while True:
        x, y = random.randint(0, row-1), random.randint(0, col-1)
        if mat[x][y] != 2:
            vacuum_pos = [x, y]
            break

    print("Sàn ban đầu:")
    for i in mat:
        print(i)

    print(f"\nBắt đầu tại: {vacuum_pos[0]}, {vacuum_pos[1]}")

    while not check_floor(mat):
        x, y = vacuum_pos[0], vacuum_pos[1]

        current_state = (x, y, tuple(tuple(r) for r in mat))
        visited.add(current_state)

        if mat[x][y] == 1:
            print(f"Bước {count+1}: Hút bụi tại {x}, {y}")
            mat[x][y] = 0
            count += 1
            continue

        moves = possible_move(x, y, mat)
        unvisited_moves = []

        for dx, dy in moves:
            new_x, new_y = x + dx, y + dy
            sim_tuple = (new_x, new_y, tuple(tuple(r) for r in mat))

            if sim_tuple not in visited:
                unvisited_moves.append((dx, dy, new_x, new_y))

        if not unvisited_moves:
            print("\nKẹt")
            break

        chosen_dx, chosen_dy, next_x, next_y = random.choice(unvisited_moves)

        vacuum_pos[0] = next_x
        vacuum_pos[1] = next_y
        count += 1

        print(f"Bước {count}: Di chuyển đến {vacuum_pos[0]}, {vacuum_pos[1]}")

    if check_floor(mat):
        print("\nHoàn thành")
    else:
        print(f"\nDừng lại ở bước: {count}")
        print("Sàn:")
        for i in mat:
            print(i)

if __name__ == "__main__":
    main()
```
